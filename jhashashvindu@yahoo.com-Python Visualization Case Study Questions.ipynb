### 6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as -
### 	Jan - Mar : Q1

### 	Apr - Jun : Q2

### 	Jul - Sep : Q3

### 	Oct - Dec : Q4   


```python
month =sales["Month"]
```

```python
month
```

```python
quarter = []
for x in month :
    if x in ["Jan","Feb","Mar"]:
        quarter.append("Q1")
    elif x in ["Apr","May","Jun"]:
        quarter.append("Q2")
    elif x in ["Jul","Aug","Sep"]:
        quarter.append("Q3")
    else:
        quarter.append("Q4")
```

```python
quarter
```

```python
sales["Qtr"]= pd.Series(quarter)
```

```python
sales.head()
```

### 7. Compare Qtr wise sales in 2015 and 2016 in a bar plot

```python
sales7
```

```python
sales7.plot(kind="bar")
plt.ylabel("Sales")
plt.title("Quarter wise sales in 2015 and 2016")
plt.show()
```

### 8. Determine the composition of Qtr wise sales in and 2016 with regards to all the Tiers in a pie chart.

```python
sales8 = sales[sales['Tier']!='Out']
sales8 = sales8.groupby(['Tier','Qtr'])['Sales2016'].sum()
sales8 = sales8.round(decimals=0)
sales8 = sales8.unstack(level=1)
sales8
```

```python
sales8.plot(kind='pie', 
            labels=['High', 'Low', 'Medium'], 
            colors=['deepskyblue', 'darkorange', 'darkgreen'],
            autopct='%1.0f%%',
            fontsize=10,
            subplots='true',
            layout=(2,2), #to get a grid
            figsize=(10,10))
plt.show()
```
